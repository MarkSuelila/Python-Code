```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

 # Load data from Excel
data = pd.read_excel(r"C:\Users\Admin\Desktop\Intel_Sales.xlsx")
df = pd.DataFrame(data)

# Total sales per product
total_sales_product = df.groupby('Intel_Product')['Unit_Sold'].sum()
print("\nTotal sales per Intel product:\n", total_sales_product)

# Total sales per day, sorted by day order
total_sales_day = df.groupby('Day')['Unit_Sold'].sum()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
total_sales_day.index = pd.CategoricalIndex(total_sales_day.index, categories=day_order, ordered=True)
total_sales_day = total_sales_day.sort_index()

print("\nTotal sales per day (sorted):\n", total_sales_day)

#Visualizing using Plot Chart
plt.figure(figsize=(8, 5))
total_sales_product.plot(kind='bar', color='#52796f')
for i, v in enumerate(total_sales_product):
    plt.text(i,v, str(v),
  ha='center',va='top', fontweight='bold' )
plt.title('Total Unit sold per Product')
plt.xlabel('Intel Processor')
plt.ylabel('Unit Sold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print('\n')

#Visualiazing using line chart
plt.figure(figsize=(8, 5))
total_sales_day.plot(kind='line',marker='o', color='#38b000')

#for loop to show the data inside
for i, v in enumerate(total_sales_day):
    plt.text(i,v, str(v),
  ha='center',va='top', fontweight='bold' )

plt.title('Total unit Sold per Day')
plt.xlabel('Days of the Week')
plt.ylabel('Unit Sold')
plt.grid(True)
plt.tight_layout()
plt.show()

print('\n')
#Pie Chat
label = df['Intel_Product'].unique()
color = ['#00b57b','#01a7c2','#52796f']
plt.pie(total_sales_product, labels=label,
autopct='%1.1f%%',startangle=90,colors=color)
plt.title("Intel Processor Total Product Sales")
plt.show()
```
